```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import QuantLib as ql
```

```python
%matplotlib inline
```

```python
today = ql.Date(27, 3, 2015)
ql.Settings.instance().setEvaluationDate(today)
cal = ql.TARGET()
fwd = ql.QuoteHandle(ql.SimpleQuote(0.03))
yts = ql.RelinkableYieldTermStructureHandle(ql.FlatForward(0,cal, fwd, ql.Actual365Fixed()))
vts = ql.RelinkableBlackVolTermStructureHandle(ql.BlackVarianceCurve(today, 
                                                                     [today+ql.Period("1y")], 
                                                                     [0.1], 
                                                                     ql.Actual365Fixed()))
vts.enableExtrapolation()
```

```python
s0 = ql.QuoteHandle(ql.SimpleQuote(100.0))
```

```python
process = ql.BlackScholesProcess(s0, yts, vts)
```

```python
dt = 1./12.
srdt = np.sqrt(dt)
tg = np.arange(0,10,dt)
```

```python
N = 1000
S = np.zeros((N,len(tg)))
for n in range(0, N):
    S[n,0] = s0.value()
    for i in range(1, len(tg)):
        S[n,i] = process.evolve(tg[i-1], S[n, i-1], dt, np.random.randn())
```

```python
for i in range(0,N):
    plt.plot(tg, S[i,])
```

Classical exposure simulation using closed formula for pricing the option on the grid
--------------------------------------------------------------------------------

```python
# Setup portfolio 
# List of Call-Option (Strike, Maturity)
portfolio = [(103, 1), (115, 10)]
V = np.zeros((len(portfolio), len(S), len(tg)))
for d in range(0, len(portfolio)):
    deal = portfolio[d]
    strike = deal[0]
    maturity = deal[1]
    payoff = ql.PlainVanillaPayoff(ql.Option.Call, deal[0])
    for i in range(0, N):
        for j in range(0, len(tg)):
            t = tg[j]
            if t > deal[1]:
                V[d,i,j] = 0
            else:
                V[d,i,j] = ql.BlackCalculator(payoff, S[i,j] * np.exp(-0.03 * (maturity-t)), np.sqrt(maturity-t) * 0.1).value()              
```

```python
for i in range(0,N):
    plt.plot(tg, V[1,i,])
```
